```python
# Importing the mammos-entity module, which provides the Entity class and specific classes for magnetic properties
import mammos_entity as me
```

```python
# Definition using mammos-entity which
me.Entity("SpontaneousMagnetization", 0.7, "MA/m", "Measured with technique ABC")
```

```python
# Shorter and more practical alternative, defining a variable Ms which holds all the information
Ms = me.Ms([150, 200, 310], "kA/m", description="Measured with technique ABC")
```

```python
# Accessing the label of Ms from the ontology, which was before automatically linked to the variable name Ms
Ms.ontology_label
```

```python
# Accessing the label of Ms with unique IRI from the ontology
Ms.ontology_label_with_iri
```

```python
# Accessing the description/definition/elucidation of Ms directly from the ontology
print(Ms.ontology.elucidation[0])
```
